# Data mined from this process can also be inserted into a csv file for further Data Analytics and Forensics

## sample code will be like

```python
from bs4 import BeautifulSoup
import requests
import csv

#insert URL of ur choice
source = request.get('Any desirable URL').text
soup = BeautifulSoup(source, 'lxml')
csv_file = open('New_CSV_file.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline','summary','video_link'])

for article in soup.find_all('article'):
    headline = article.h2.a.text
    print(headline)
    
    summary = article.find('div',class_='entry-content').p.text
    print(summary)
    
    try:
        vid_src = article.find('iframe', class_='youtube-player')['src']
        vid_id = vid_src.split('/')[4]
        vid_id = vid_id.split('?')[0]
        
        yt_link = f'https://youtube.com/watch?v={vid_id}'
    except Exception as e:
        yt_link = None
        
    print(yt_link)
    
    csv_writer.writerow([headline, summary, yt_link])
    
csv_file.close()
```
